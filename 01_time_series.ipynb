- 🗓️ First, we convert the **Closing Date** to a datetime format. ✅

```python

# Strip whitespace from column names
df.columns = df.columns.str.strip()

# Convert closing date to datetime
df['Closing Date'] = pd.to_datetime(df['Closing Date'], errors='coerce')

# Drop rows with invalid dates
df = df.dropna(subset=['Closing Date'])

# Show first few rows
df.head()
```

```python
# Set date as index
df = df.set_index('Closing Date')

# Resample monthly and count number of closures
monthly_failures = df.resample('M').size()

# Show result
monthly_failures.head()
```

- 📊 Let's Decompose our Time Series into **Trend**, **Seasonality**, and **Residuals**. 📈📉

```python
# Decompose the time series
decomposition = seasonal_decompose(monthly_failures, model='additive', period=12)
# Create a figure for the decomposition plot
plt.figure(figsize=(12, 8))
# Plot observed data
plt.subplot(4, 1, 1)
plt.plot(decomposition.observed, label='Observed', color='blue')
plt.title('Observed', fontsize=14)
plt.legend()
# Plot trend with a custom color
plt.subplot(4, 1, 2)
plt.plot(decomposition.trend, label='Trend', color='green')
plt.title('Trend', fontsize=14)
plt.legend()
# Plot seasonality with a custom color
plt.subplot(4, 1, 3)
plt.plot(decomposition.seasonal, label='Seasonality', color='orange')
plt.title('Seasonality', fontsize=14)
plt.legend()
# Plot residuals with a custom color
plt.subplot(4, 1, 4)
plt.plot(decomposition.resid, label='Residuals', color='red')
plt.title('Residuals', fontsize=14)
plt.legend()
# Adjust layout for better spacing
plt.tight_layout()
# Show the plot
plt.show()
```

**Seasonality** 🔄 (yellow) displays regular, repeating oscillations, showing consistent periodic patterns 🔁.  
**Residuals** ⚠️ (red) highlight irregular variations, with the biggest deviations during the 2010-2012 spike 📅.

## 📈 Let's Apply the ARIMA Model 🚀

- 📊 We take **d=0** because our data is **stationary** ✅, and **p=1** and **q=1** by analyzing the **PACF** and **ACF** plots. Hence, the order of **(p, d, q)** is **(1, 0, 1)**. 🚀

```python
train = monthly_failures[:-12]
test = monthly_failures[-12:]

model = ARIMA(train, order=(1,0,1))
model_fit = model.fit()
forecast = model_fit.forecast(steps=12)

# Plot results
plt.plot(train, label='Train')
plt.plot(test, label='Test')
plt.plot(forecast, label='Forecast', color='orange')
plt.legend()
plt.show()
```

- 🌟 "Based on ARIMA (1,0,1) forecasting over the monthly bank failure data 📊, the next 12 months 📅 are expected to show minimal to negligible bank failures in USA🏦, indicating a period of financial stability 💹 in the banking sector." 🎉

```python

rmse = np.sqrt(mean_squared_error(test, forecast))
print(f"RMSE: {rmse:.2f}")
```

- 📊 Let's apply the **SARIMA model** on the same data to correct the residual noise, which is minimal. ✅

```python

model = SARIMAX(train,
                order=(1,0,1),
                seasonal_order=(1,0,1,12),
                enforce_stationarity=False,
                enforce_invertibility=False)
results = model.fit(disp=False)
```

- 📊 Let's plot the diagnostics in our residual. 🔍

```python

results.plot_diagnostics(figsize=(15, 8))
plt.show()
```

- 📊 Let's check the residuals with a histogram. 📈

```python

residuals = results.resid
plt.figure(figsize=(10,4))
sns.lineplot(x=train.index, y=residuals)
plt.title("Residuals Over Time")
plt.axhline(0, color='red', linestyle='--')
plt.show()

# Histogram of residuals
sns.histplot(residuals, kde=True)
plt.title("Distribution of Residuals")
plt.show()
```

- 📊 Let's Forecast our data using the **SARIMAX Model** 🚀 and also we check our **RMSE** ✅.

```python

forecast = results.get_forecast(steps=12)
forecast_ci = forecast.conf_int()
forecast_values = forecast.predicted_mean

# Plot
plt.figure(figsize=(12,6))
plt.plot(monthly_failures, label='Observed')
plt.plot(forecast_values, label='Forecast')
plt.fill_between(forecast_ci.index, 
                 forecast_ci.iloc[:, 0], 
                 forecast_ci.iloc[:, 1], color='pink', alpha=0.3)
plt.legend()
plt.title("Forecast with SARIMAX")
plt.show()

rmse = np.sqrt(mean_squared_error(test, forecast_values))  # Use forecast_values instead of forecast
print("RMSE on Test Set:", round(rmse, 4))
```
